```python
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
```

```python
basic_array = [67, 54, 59, 72, 68, 37, 58, 38, 41, 54, 45, 64, 57, 50, 55, 53, 65, 45, 53, 38]

plt.plot(basic_array)
plt.show()

pprint(basic_array)
```

-----------
## 1. Sum (Σ)
The sum of a set of values is the addition of all the values in the set.
- **Formula:** $\Sigma x_i$

```python
def sum (values):
    values_sum = 0

    for v in values:
        values_sum = values_sum + v
    
    return values_sum
```

```python
# Sample

print(f"Sum: {sum(basic_array)}")
```

-------
## 2. Mean (Average)
The mean is the sum of all the values divided by the number of values.
- **Explanation:** The mean provides a measure of the central tendency of the data.
- **Formula:** $\bar{x} = \frac{\Sigma x_i}{n}$

```python
def average (values):
    #getting the length of values, the number of items.
    number = values.__len__()
    return sum(values) / number
```

```python
# Sample

print(f"Number of values: {basic_array.__len__()}")
print(f"Average: {average(basic_array)}")
```

-----------
## 3. Deviations from the Mean
Each value's deviation from the mean is calculated by subtracting the mean from the value.
- **Explanation:** Deviations indicate how far each value is from the mean.
- **Formula:** $x_i - \bar{x}$

```python
def deviations(values):
    avg = average(values)
    c_deviations = []
    
    #determining the difference between the item from the mean (deviation)
    for v in values:
        c_deviations.append(v - avg)

    return c_deviations
```

```python
# Sample

r_deviations = deviations(basic_array)

plt.plot(r_deviations)
plt.title("Basic Array Deviations")
plt.show()

pprint(r_deviations)
```

-----------
## 4. Squared Deviations
Squaring the deviations eliminates negative values and gives more weight to larger deviations.
- **Explanation:** Squared deviations are used to measure the variability in the data.
- **Formula:** $(x_i - \bar{x})^2$

```python
def sqr_deviations(values):
    c_deviations = deviations(values)
    sqr_deviations = []
    for d in c_deviations:
        sqr_deviations.append(d**2)

    return sqr_deviations
```

```python
# Sample

r_sqr_deviations = sqr_deviations(basic_array)

plt.plot(r_sqr_deviations)
plt.title("Basic Array Square Deviations")
plt.show()

pprint(r_sqr_deviations)
```

----
## 5. Sum of Squared Deviations
The sum of squared deviations is the total of all the squared deviations.
- **Formula:** $\Sigma (x_i - \bar{x})^2$

```python
def sum_sqr_deviations(values):
    return sum(sqr_deviations(values))
```

```python
# Sample

print(f"Sum of Square Deviations: {sum_sqr_deviations(basic_array)}")
```

---
## 6. Variance (Population)
The variance is the average of the squared deviations for the entire population.
- **Explanation:** Variance measures the spread of the data points in a population.
- **Formula:** $\sigma^2 = \frac{\Sigma (x_i - \bar{x})^2}{N}$

```python
def variance(values):
    number = values.__len__()
    return sum_sqr_deviations(values) / number
```

```python
# Sample

print(f"Variance: {variance(basic_array)}")
```

---
## 7. Variance (Sample)
For a sample, the variance is calculated by dividing the sum of squared deviations by the number of values minus one.
- **Explanation:** The sample variance corrects for bias in the estimation of the population variance.
- **Formula:** $s^2 = \frac{\Sigma (x_i - \bar{x})^2}{n-1}$

```python
def variance_sample(values):
    number = values.__len__() - 1
    return sum_sqr_deviations(values) / number
```

```python
# Sample

print(f"Sample Variance: {variance_sample(basic_array)}")
```

---
## 8. Standard Deviation (Population)
The standard deviation is the square root of the variance for the population.
- **Explanation:** The standard deviation provides a measure of the spread of the data points around the mean.
- **Formula:** $\sigma = \sqrt{\sigma^2} = \sqrt{\frac{\Sigma (x_i - \bar{x})^2}{N}}$

```python
def standard_deviation(values):
    return variance(values) ** 0.5
```

```python
# Sample

print(f"Standard Deviation: {standard_deviation(basic_array)}")
```

---
## 9. Standard Deviation (Sample)
The standard deviation for a sample is the square root of the sample variance.
- **Explanation:** The sample standard deviation estimates the spread of the data points in the population based on the sample.
- **Formula:** $s = \sqrt{s^2} = \sqrt{\frac{\Sigma (x_i - \bar{x})^2}{n-1}}$

```python
def standard_deviation_sample(values):
    return variance_sample(values) ** 0.5
```

```python
# Sample

print(f"Sample Standard Deviation: {standard_deviation_sample(basic_array)}")
```

## Histogram Sample
A histogram provides a visual representation of the distribution of the data. Overlaying the mean and standard deviations can highlight the spread and central tendency.

```python
# Generate histogram
plt.hist(basic_array, bins=10, edgecolor='black', alpha=0.7)

# Plot mean and standard deviations
mean_value = average(basic_array)
sample_std_dev = standard_deviation_sample(basic_array)

plt.axvline(mean_value, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(mean_value + sample_std_dev, color='g', linestyle='dashed', linewidth=1, label='+1 SD')
plt.axvline(mean_value - sample_std_dev, color='g', linestyle='dashed', linewidth=1, label='-1 SD')

plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

---
## Box Plot (Box and Whisker Plot)

A box plot displays the median, quartiles, and outliers. The spread of the whiskers gives a visual indication of the spread of the data, which is related to the standard deviation.

```python
plt.boxplot(basic_array, vert=False)
plt.title('Box Plot of Data')
plt.xlabel('Value')
plt.show()
```

---
## Scatter Plot with Standard Deviation Bands:

If you have a time series or sequential data, you can plot the values and overlay bands representing one or two standard deviations from the mean.

```python
# Calculate mean and standard deviation for the plot
mean_value = np.mean(basic_array)
sample_std_dev = np.std(basic_array, ddof=1)

# Scatter plot with a thin line linking the points
plt.scatter(range(len(basic_array)), basic_array, color='b')
plt.plot(range(len(basic_array)), basic_array, color='b', linewidth=0.5)  # Thin line linking points

# Mean and standard deviation lines
plt.axhline(mean_value, color='r', linestyle='--', label='Mean')
plt.axhline(mean_value + sample_std_dev, color='g', linestyle='--', label='+1 SD')
plt.axhline(mean_value - sample_std_dev, color='g', linestyle='--', label='-1 SD')

# Fill between one standard deviation
plt.fill_between(range(len(basic_array)), mean_value + sample_std_dev, mean_value - sample_std_dev, color='g', alpha=0.2)

# Titles and labels
plt.title('Scatter Plot with Standard Deviation Bands')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
